Ques : Are customers with lower watch hours more likely to churn? 



==>Customers with fewer watch hours are more likely to leave.
 (Because there is a strong negative correlation between watch_hours and churn.)

==>Users who spend more time watching content tend to remain subscribed.

==>Users who churned had significantly lower average watch time compared to those who stayed.

==>Watch hours can serve as an early indicator to predict potential churn.

==>Retention strategies should focus on increasing engagement (watch hours) to reduce churn.

==>There is a moderate-to-strong negative correlation (-0.65) between watch hours and churn.

==>The most active users (top 25% by watch hours) experienced the lowest churn rate.

Q) Do Premium users churn more or less than Basic users? 


==>Premium users show the lowest churn rate, indicating better satisfaction or perceived value in the Premium plan.

==>There is a clear inverse relationship between subscription tier and churn  as subscription level increases, churn rate decreases.

==>Despite higher cost, Premium users are more loyal, suggesting that investing in higher-tier plans might improve retention.

==>Basic plan may lack features or value, leading to user dissatisfaction and eventual churn.==>

==>Standard and Premium users have more balanced churn vs non-churn proportions, while Basic users have much higher churn volume than retention.

==>Targeted retention efforts like offering feature trials or personalized offers to Basic users could significantly reduce churn.

```python
# Q)   Which region has the highest average watch_hours?
avg_watch_by_region = df.groupby('region')['watch_hours'].mean()
avg_watch_by_region = avg_watch_by_region.sort_values(ascending=False)
print(avg_watch_by_region)

```

```python
avg_watch_by_region.plot(kind='bar', color='skyblue')
```

==>North America leads all regions in average watch hours, indicating a more engaged or binge-watching audience.

==>South America and Oceania follow closely, suggesting strong user engagement in these regions as well.

==>Asia has the lowest average watch hours ( 11.28), which could indicate lower engagement or content relevance in that region.

==>The difference in average watch time across regions is not drastic, but still notable enough to guide content and marketing strategies.

==>Regions with higher average watch hours may respond better to content recommendations and upselling offers.

==>Asia and Europe, having relatively lower watch hours, might benefit from localized or more targeted content.

==>Platform performance, content library, or streaming accessibility might be influencing watch behavior by region.
